```python
# largest element in list with index
l = [12,567, 43, 235, 347, 568, 45, 7]

largest = l[0]
index = 0

for i in range(len(l)):
    if l[i] > largest:
        largest = l[i]
        index = i

print(f"your largest number is: {largest} at index: {index}")       
```

```python
# find second largest element
l = [12,16,13,19,17]
largest = l[0]

for i in l: 
    if i > largest:
        sec_largest = largest
        largest = i
    elif i > sec_largest:
        sec_largest = i
print(sec_largest, largest)        

```

```python
# check list is sorted or not
l = [12, 13, 18, 15,16]

for i in range(len(l) - 1):
    if l[i] < l[i + 1]:
        continue
    else:
        print("your list is not sorted")
        break
else:
    print("your list is sorted")    
```

```python
# Dictionary
# key: value pairs
# mutable

info ={
    "name":"diksha",
    "age": 22,
    "cgpa": 7.4,
    "subjects":["python", "machine learning","DBMS"],
    3.14: "PI"
}

info["cgpa"] = 7.4

print(info["cgpa"])
```

```python
# Dictionary methods
# . d.keys(), .d.values(), .d.items(), .d.get(val), .d.update(new_item)

#d.keys()  # returns all keys
info ={
    "name":"diksha",
    "age": 22,
    "cgpa": 7.4,
    "subjects":["python", "machine learning","DBMS"],
    3.14: "PI"
}
dict_keys = list(info.keys())
print(dict_keys)

```
