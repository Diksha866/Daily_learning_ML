```python
# 5. Challenge:
# Create a list of even numbers from 2 to 20 using list repetition or a loop (if already covered).
# Store multiple student records in a nested list (e.g., ["Alice", 90])

# Part 1: Even numbers from 2 to 20 using a loop
even_numbers = []
for num in range(2, 21, 2):  # start=2, end=20, step=2
    even_numbers.append(num)
print("Even numbers (using loop):", even_numbers)

# Alternative using list comprehension (more Pythonic)
even_numbers_alt = [num for num in range(2, 21, 2)]
print("Even numbers (list comprehension):", even_numbers_alt)

print()

# Part 2: Store multiple student records in a nested list
students = [
    ["Alice", 90],
    ["Bob", 85],
    ["Charlie", 92],
    ["Diana", 88],
    ["Eve", 95]
]

print("Student Records:")
print(students)

print()

# Access specific records
print("First student:", students[0])  # ["Alice", 90]
print("Alice's score:", students[0][1])  # 90

print()

# Iterate through all student records
print("All students with their scores:")
for student in students:
    print(f"{student[0]}: {student[1]}")
```
