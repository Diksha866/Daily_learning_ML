```python
# Dictionary -> A dictionary in python is an unordered, mutable collection of key-value pairs, 
# it is one of  the  most powerfull built-in  data structures, used to map unique keys to values.

person = {
  "name": "alica",
  "age": 23,
  "profession": "student"
}
print(person)
```

```python
# Accessing and Modifying Values
print(person["name"])  # access value
person["age"] = 31 # modify value
person["profession"] = "teaching" # Add new key-value pair
```
