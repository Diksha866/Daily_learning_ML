```python
# if-elif-else
temp = 23
if temp >= 30:
  print("hot")
else:
  print("cold")  
```

```python
#elif
temp = 23
if temp >= 30:
  print("hot")
elif temp<30 and temp > 18:
  print("pleasant")  
else:
  print("cold") 
```

```python
# for loop 
# a for loop is used to iterate over sequence (like list, strings, dictionaries).
name = ["apple", "banana", "kiwi"]
for item in name:
  if len(item)<=5:
     print("short")
  else:
     print("long")   
  
```
